# Inverting river sediments for source-region geochemistry 

In this document we develop a scheme to invert compositions of higher-order river sediments to derive a map of upstream geochemistry. We do this by minimising the misfit between the predicted river geochemistry and the observed ones for samples gathered in the Cairngorms region of the UK. The spatial resolution of our inverse technique and the uniqueness of the misfit space is explored. Finally, we use external optimisation libraries to generate a 'best-fitting' solution for our given computational abilities. This best-fitting solution is compared to the 'real' source-region geochemistry as derived by geochemical surveying. Hence we can externally validate our method.

# Initialise Model Space

## Loading and preprocessing data 

### Topography 

Topographic data for studied region is taken from `SRTM1s` dataset, gridded to 200 m $\times$ 200 m squares following a cylindrical equal-area projection centred on the study area. 

```python
zr_nc=netCDF4.Dataset('input_dir/topo_CG.nc')
zr_ma = zr_nc['z'][:,:]
mg = RasterModelGrid(zr_ma.shape,xy_spacing=(200,200))

landmask_nc=netCDF4.Dataset('input_dir/landmask.nc')
landmask = landmask_nc['z'][:,:].data.astype(float)

zr = mg.add_zeros('node', 'topographic__elevation')
zr += np.reshape(zr_ma.data.astype(float),zr.shape)
dx = np.loadtxt("input_dir/dx_CG.dat")

# Set up the boundary conditions on the square grid
mg.set_fixed_value_boundaries_at_grid_edges(True,True,True,True)

flat_shape = zr.shape # a tuple to flatten arrays [number of nodes long]
full_shape = mg.shape # the full shape of the grid [rows, columns]
```

This raw data needs to have the sinks filled so that the flow-routing algorithm can perform correctly. This can take some time. 

```python
sfb = SinkFillerBarnes(mg, method='D8',fill_flat=False)
sfb.run_one_step()
```

The final DEM is shown here:

```python
plt.figure()
plt.title("Topographic Elevation")

plt.imshow(zr.reshape(full_shape)+landmask,cmap='cubehelix',origin='lower') 
plt.xlabel('Horizontal grid cells')
plt.ylabel('Vertical grid cells')
plt.show()
```

To process the downstream samples better we need to produce a model of the drainage network and initiate the incision and cumulative flux grids.

```python
# instantiate the flow routing:

frr = FlowAccumulator(
    mg,
    'topographic__elevation',
    flow_director = 'FlowDirectorD8')
frr.run_one_step()  # flow routing

zrold = zr[mg.nodes] # pre-incision elevations

# instantiate the stream power law
k=3.62
m=0.35
n=1
spr = FastscapeEroder(mg, K_sp=k, m_sp=m, n_sp=n)

dt=0.1# Timestep in Ma
spr.run_one_step(dt)    # erosion: stream power
zrnew = zr[mg.nodes]    # post-incision elevations
incise = zrold - zrnew  # incision per cell
cell_area = mg.dx * mg.dy
qs = incise*cell_area # Volume sediment produced per cell
qsflat = qs.ravel()     # flatten qs for flow routing calculation

a,q = find_drainage_area_and_discharge(mg.at_node['flow__upstream_node_order'], mg.at_node['flow__receiver_node'],runoff = qsflat) 
# q is cumulative flux for SPL model
a, q_homo_incis = find_drainage_area_and_discharge(mg.at_node['flow__upstream_node_order'], mg.at_node['flow__receiver_node']) # a is number of nodes
# q is cumulative flux for homogeneous incision

area = mg.at_node['drainage_area']
np.amax(area)
area_threshold = 25 #float(sys.argv[1]) #25 km2
is_drainage = area > (area_threshold*1000000) #km2 to m2
mg.add_field('node','channels',is_drainage,noclobber=False)

plt.rcParams["figure.figsize"] = (12,8)
plt.figure()
plt.imshow(is_drainage.reshape(full_shape)+landmask,origin='lower') 
plt.xlabel('Horizontal grid cells')
plt.ylabel('Vertical grid cells')

plt.title("Channels > 25 km2")
plt.show()
```


### Higher-order river geochemistry  

The above plot shows the true sample localities as measured by GPS at the sample sites. However, we need to snap these locations to the nearest channel location to which they belong so that we can access the model grid at the channels. To 'snap' these points to the model grid we simply identify the nearest cell which is marked as a 'channel' cell. However, slight discrepancies between the true drainage network and the model drainage network, plus error in GPS locations, mean that for some localities we have to give them a little 'nudge' to make them snap to the true locality.

```python

sample_data = np.loadtxt('input_dir/samples.dat',dtype=str) # [sample #, locality #, x, y]
sample_locs = sample_data[:,2:4].astype(np.float)

channel_xy = np.flip(np.transpose(np.where(is_drainage.reshape(mg.shape))),axis=1)*200 # xy coordinates of channels

nudge = np.zeros(sample_locs.shape) # initiate nudge array

nudge[np.where(sample_data[:,1]=='2'),:] = [-200,-200] # Nudging locality 2 to the SW 
nudge[np.where(sample_data[:,1]=='6'),:] = [-600,0] # Nudging locality 6 to the W 
nudge[np.where(sample_data[:,1]=='13'),:] = [-200,0] # Nudging locality 13 to the W 
nudge[np.where(sample_data[:,1]=='19'),:] = [0,400] # Nudging locality 19 to the N 
nudge[np.where(sample_data[:,1]=='20'),:] = [-400,-400] # Nudging locality 20 to the SW 
nudge[np.where(sample_data[:,1]=='23'),:] = [-400,0] # Nudging locality 23 to the W 
nudge[np.where(sample_data[:,1]=='29'),:] = [-300,0] # Nudging locality 29 to the W 
nudge[np.where(sample_data[:,1]=='53'),:] = [-600,-600] # Nudging locality 53 to the SW 
nudge[np.where(sample_data[:,1]=='54'),:] = [0,600] # Nudging locality 54 to the N 
nudge[np.where(sample_data[:,1]=='56'),:] = [-200,-200] # Nudging locality 56 to the SW 
nudge[np.where(sample_data[:,1]=='57'),:] = [-1200,-1200] # Nudging locality 57 to the SW 

nudged_locs = sample_locs + nudge # Apply the nudges

# Fit the data to the nearest channel node

fitted_locs = np.zeros(sample_locs.shape) # Initialise snapped locality array
for i in np.arange(nudged_locs.shape[0]):
    sample = nudged_locs[i,:]
    diffs = channel_xy-sample
    distances = np.linalg.norm(diffs,axis=1)
    shortest = np.amin(distances)
    fitted_locs[i,:] = channel_xy[np.where(distances==shortest)]    
    
```

## Compare observations and model predictions

Now we check that our model is correctly intialised by testing its ability to make sensible model predictions when provided with the true G-BASE data. 
The below function takes as input an element string and then plots up a cross plot of model predictions and prints the RMS misfit. Note that there are _minor_ differences to the previous RMS values (from the manuscript) because the way we are snapping to the model drainage grid is slightly different. We test it for both the stream power erosion model and the simpler homogeneous incision model. 


```python
%matplotlib inline

def abline(slope, intercept):
    # Auxiliary function
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--',c='gray')
    
def test_model(elem): 
    """ Tests model against GBASE input """
    comp_nc=netCDF4.Dataset('input_dir/input_chems/CG_log_'+elem+'.nc')
    comp_ma = comp_nc['z'][:,:]
    comp_bedrock = 10**comp_ma.data.astype(float).reshape(flat_shape)
    
    # For stream power model 
#     a,sed_comp = find_drainage_area_and_discharge(mg.at_node['flow__upstream_node_order'], mg.at_node['flow__receiver_node'],runoff = comp_bedrock*qsflat)
#     with np.errstate(divide='ignore'):
#         with np.errstate(invalid='ignore'):
#             sed_comp_norm = sed_comp/q
#     sed_comp_norm[q==0] = comp_bedrock[q==0]   

    # For homogeneous incision 
    a, sed_comp = find_drainage_area_and_discharge(mg.at_node['flow__upstream_node_order'], mg.at_node['flow__receiver_node'],runoff = comp_bedrock)  # composition but homogeneous erosion
    with np.errstate(divide='ignore'):
        with np.errstate(invalid='ignore'):     
            sed_comp_norm = sed_comp /q_homo_incis
    sed_comp_norm[q_homo_incis==0] = comp_bedrock[q_homo_incis==0]

    preds = sed_comp_norm[loc_nodes]
    obs = obs_data[elem]
                             
    rms = np.sqrt(np.mean((np.log10(obs) - np.log10(preds))**2))
    print("RMS misfit for", elem," =", rms)
    
    plt.figure()    
    plt.scatter(x=preds,y= obs)
    # Add regression line
    plt.plot(np.unique(preds), 10**np.poly1d(np.polyfit(np.log10(preds), np.log10(obs), 1))(np.unique(np.log10(preds))),c='k')
    plt.xlabel('Predicted '+elem+', mg/kg')
    plt.ylabel('Observed '+elem+', mg/kg')
    ax = plt.gca()
    abline(1,0)
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.axis('equal')
    
    
test_model('Mg')        
```

Grey line is the target 1:1 line and the black line is the regression through the data. These tests suggest that our model is initialised correctly and we can now investigate changing the input systematically.

# Initialising an inversion
## Defining a cost-function

Now we define a number of functions that turn a geochemical input map into a measure of data misfit. These functions will be used to identify the 'best-fitting' solutions.  

```python
def cost_rms_spl(bdrck_arr,elem): 
    """Returns RMS misfit for a given bedrock input array, calculating predictions using the SPL"""
    a,sed_comp = find_drainage_area_and_discharge(mg.at_node['flow__upstream_node_order'], mg.at_node['flow__receiver_node'],runoff = bdrck_arr*qsflat)
    with np.errstate(divide='ignore'):
        with np.errstate(invalid='ignore'):
            sed_comp_norm = sed_comp/q
    sed_comp_norm[q==0] = bdrck_arr[q==0]   
    
    preds = sed_comp_norm[loc_nodes]
    obs = obs_data[elem]
                             
    rms = np.sqrt(np.mean((np.log10(obs) - np.log10(preds))**2))
    return(rms)

def cost_rms_homo(bdrck_arr,elem):
    """Returns RMS misfit for a given bedrock input array, calculating predictions using the homogenous incision"""
    a, sed_comp = find_drainage_area_and_discharge(mg.at_node['flow__upstream_node_order'], mg.at_node['flow__receiver_node'],runoff = bdrck_arr)  # composition but homogeneous erosion
    sed_comp_norm = sed_comp/q_homo_incis           
    rms = np.sqrt(np.mean((np.log10(obs_data[elem]) - np.log10(sed_comp_norm[loc_nodes]))**2))
    return(rms)
    
def cost_homo(bdrck_arr,elem):
    """Returns L2norm data misfit for a given bedrock input array (mgkg), calculating predictions using the homogenous incision"""
    a, sed_comp = find_drainage_area_and_discharge(mg.at_node['flow__upstream_node_order'], mg.at_node['flow__receiver_node'],runoff = bdrck_arr)  # composition but homogeneous erosion
    sed_comp_norm = sed_comp/q_homo_incis           
    l2norm = np.linalg.norm(np.log10(obs_data[elem]) - np.log10(sed_comp_norm[loc_nodes]))
    return(l2norm)    
    
Mg_comp_nc=netCDF4.Dataset('input_dir/input_chems/CG_log_Mg.nc')
Mg_comp_ma = Mg_comp_nc['z'][:,:]
Mg_comp_bedrock = 10**Mg_comp_ma.data.astype(float).reshape(flat_shape)

U_comp_nc=netCDF4.Dataset('input_dir/input_chems/CG_log_U.nc')
U_comp_ma = U_comp_nc['z'][:,:]
U_comp_bedrock = 10**U_comp_ma.data.astype(float).reshape(flat_shape)


cost_homo(Mg_comp_bedrock,'Mg')
```

## Creating synthetic source region maps 

To invert for the best-fitting source-region geochemistry we need a way to automatically generate geochemical maps. We do this by define a grid of 'sticks' of variable height. To create a continuous map from these sticks we interpolate linearly, although we can also use a cubic or quintic spline. The number of sticks in the x and y direction are hyper-parameters than can be modified. The variables which we vary to minimise the misfit are the stick heights.

```python
n_stick_x = 10
n_stick_y = 10

width = mg.shape[1] # number of x cells in topographic model grid
height = mg.shape[0] # number of y cells in topographic model grid

X_new = np.arange(0,width,1) # x-points where interpoland will be evaluated
Y_new = np.arange(0,height,1) # y-points where interpoland will be evaluated
  
x_sticks = np.linspace(0,width,n_stick_x) # location in model x-dir of the sticks
y_sticks = np.linspace(0,height,n_stick_y) # location in model y-dir of the sticks

z = np.random.random((n_stick_y,n_stick_x)) # Generate a random 2D array

def z_to_grd(z_arr):
    f=sp.interpolate.interp2d(x_sticks,y_sticks,z_arr,kind='linear') # can also use 'cubic' or 'quintic' but slower
    interpd = f(X_new,Y_new) # Evaluate interpoland on the defined model grid
    return(interpd)

interpoland = z_to_grd(z)
stick_mesh_xx, stick_mesh_yy = np.meshgrid(x_sticks,y_sticks)

plt.figure()
plt.title("Example of interpolating between stick heights")
plt.ylabel("Vertical grid cells")
plt.xlabel("Horizontal grid cells")
plt.imshow(interpoland,origin='lower')
plt.scatter(x=stick_mesh_xx,y=stick_mesh_yy,c='r',marker='x')
```

The function `z_to_grd` takes an array of stick heights and linearly interpolates between them to generate a continous grid at the resolution of the model landscape grid. The above plot shows the output for a series of random stick heights. The stick 'locations' are marked by red crosses. The time-cost of this function is 8 ms for a linear interpolation, 17ms for cubic and 30.6 ms for quintic.

These grids can now be fed into the cost function.  For example:

```python
cost_homo(interpoland.reshape(flat_shape),'Mg')
```

### Removing redundant 'sticks'

Thus the parameter space to invert over is simply the array of stick heights 'z'. But, this parameter space includes sticks for which our samples are insensitive to because they are not contained within the studied drainage areas. These 'sticks' need to be identified and set to a constant value prior to inversion (e.g. -1). First we identify the 'active' parts of the grid by identifying the catchment areas of our studied rivers.

```python
lowest_tay_sample = loc_nodes[sample_data[:,1]=='55'] # Locality 55
tay_catchment = get_watershed_mask(mg,lowest_tay_sample) # extract upstream area of most downstream tay sample 

lowest_dee_sample = loc_nodes[sample_data[:,1]=='7'] # Locality 7. 
dee_catchment = get_watershed_mask(mg,lowest_dee_sample)

lowest_don_sample = loc_nodes[sample_data[:,1]=='34'] # Locality 34. 
don_catchment = get_watershed_mask(mg,lowest_don_sample)

lowest_spey_sample = loc_nodes[sample_data[:,1]=='28'] # Locality 28. 
spey_catchment = get_watershed_mask(mg,lowest_spey_sample)

lowest_deveron_sample = loc_nodes[sample_data[:,1]=='40'] # Locality 28. 
deveron_catchment = get_watershed_mask(mg,lowest_deveron_sample)

active_area = tay_catchment | dee_catchment | don_catchment | spey_catchment | deveron_catchment

coverage_mask = np.invert(active_area.reshape(full_shape))  # This is a bool array which is used to mask regions with no coverage
coverage_nan_mask = np.zeros(coverage_mask.shape)
coverage_nan_mask[np.where(coverage_mask)] = 'nan'
mg_active_cell_bool = np.where(np.invert(coverage_mask))

plt.figure()
plt.title("Coverage relative to inversion grid cells (including coastline)")
plt.ylabel("Vertical model grid cells")
plt.xlabel("Horizontal model grid cells")
plt.imshow(active_area.reshape(full_shape)+landmask,origin='lower')
plt.scatter(x=stick_mesh_xx,y=stick_mesh_yy,c='r',marker='x')
```

The above figure shows the active area in yellow, overlain with the location of the nodes. As can be seen, many of these nodes are redundant. To identify these takes two steps. 

The first step is to identify the inactive *cells*, i.e., the inter-node spaces which contain no active area. These are the completely purple cells in the above image. This is performed by the following chunk which returns a 2D boolean array showing which cells are active or not.  

```python
active_area_full= active_area.reshape(full_shape)
blocked_active = np.zeros((n_stick_y-1,n_stick_x-1))
x_step=width/(n_stick_x-1)
y_step=height/(n_stick_y-1)

for i in np.arange(n_stick_x-1):
    for j in np.arange(n_stick_y-1):
        start_x=int(np.ceil(i*x_step))
        end_x=int(np.floor((i+1)*x_step))
        start_y=int(np.ceil(j*y_step))
        end_y=int(np.floor((j+1)*y_step))
        blocked_active[j,i] = np.any(active_area_full[start_y:end_y,start_x:end_x])    
plt.imshow(blocked_active,origin='lower')
plt.ylabel("Vertical inversion grid cells")
plt.xlabel("Horizontal inversion grid cells")
plt.title("Active Cells")
plt.show()
```

Now that we have indentified the active *cells* we want to identify the sticks which are *only* surrounded by inactive cells. These are the *inactive* cells we want to exclude from our inversion.

```python
blocked_active = blocked_active.astype(bool)
active_sticks = np.zeros((n_stick_y,n_stick_x)).astype(bool)
for i in np.arange(n_stick_x):
    for j in np.arange(n_stick_y):
        if((i,j)==(0,0)): # Top left stick
            active_sticks[j,i] = blocked_active[j,i]
        elif ((i,j)==(n_stick_x-1,0)): # Top right stick
            active_sticks[j,i] = blocked_active[j,i-1] 
        elif ((i,j)==(0,n_stick_y-1)): # Bottom left
            active_sticks[j,i] = blocked_active[j-1,i]
        elif ((i,j)==(n_stick_x-1,n_stick_y-1)): # Bottom right
            active_sticks[j,i] = blocked_active[j-1,i-1]
        elif (j==0): # Top row
            active_sticks[j,i] = blocked_active[j,i] | blocked_active[j,i-1]
        elif (j==n_stick_y-1): # Bottom row
            active_sticks[j,i] = blocked_active[j-1,i] | blocked_active[j-1,i-1]
        elif (i==0): # left column
            active_sticks[j,i] = blocked_active[j-1,i] | blocked_active[j,i]    
        elif (i==n_stick_x-1): # right column
            active_sticks[j,i] = blocked_active[j-1,i-1] | blocked_active[j,i-1]   
        else: # centre cells
            active_sticks[j,i] = blocked_active[j-1,i-1] | blocked_active[j,i-1] | blocked_active[j,i] | blocked_active[j-1,i]  

            
plt.imshow(active_area.reshape(full_shape)+landmask,origin='lower')
plt.ylabel("Vertical model grid cells")
plt.xlabel("Horizontal model grid cells")
plt.title("Active (red) vs inactive (blue) sticks")
plt.scatter(x=stick_mesh_xx,y=stick_mesh_yy,c='b',marker='x')
plt.scatter(np.where(active_sticks)[1]*x_step,np.where(active_sticks)[0]*y_step,c='r',marker='x')  
plt.show()
```

The above figure shows the results of our algorithm for identifying active and inactive nodes. These are excluded from the inversion.

### Defining source region priors 

Now we define a 'prior' estimate of what the concentration of each element should be. This value could be used to intitiate an inversion closer to the minimum (improving convergence) or to regularize the solution if the misfit space is unhelpful.

There are a number of options to define a prior concentration, if we make the assumption that concentration is homogenous everywhere. One option is to take the average concentration of that element in the upper continental crust (note this would not incorporate the effect of weathering). A second option is to take the (arithmetic) average of all the higher-order samples. A final option is to take the average of the concentrations at the mouths of the river weighted by the upstream area. These options (based on geological intuition) are likely increasingly accurate representations of the 'target' map. These priors are defined below. None of these are particularly tight constraints, nor require any information that is not readily available (e.g. UCC) or calculated from data already gathered for the inversion.

## Initiating inversions

Now we redefine a `initiate_inversion` function which sets up a grid and returns a list of the active nodes. 

```python
def initiate_inversion(nx,ny,plot=True,init='logit',upp=999999,low=1):
    """ Initiates a model grid with 'nx' sticks along the x-axis and 
    'ny' sticks along y-axis. Defines global variables about the grid (e.g. spacing). 
    Automatically works out which of these nodes are sampled by our data-coverage.
    
    Can initiate the grid with random values either equally distributed in logistic space
    (default) or linear. Random values are distributed between the upp and low limits.
    """
    
    # Defines global variables based on grid resolution. Some of these are used by z_to_grd()
    global x_sticks, y_sticks,x_step,y_step,n_stick_x,n_stick_y
    n_stick_x = nx
    n_stick_y = ny

    width = mg.shape[1] # number of x cells in topographic model grid
    height = mg.shape[0] # number of y cells in topographic model grid

    x_sticks = np.linspace(0,width,n_stick_x) # location in model x-dir of the sticks
    y_sticks = np.linspace(0,height,n_stick_y) # location in model y-dir of the sticks
    
    active_area_full= active_area.reshape(full_shape)
    x_step=width/(n_stick_x-1)
    y_step=height/(n_stick_y-1)

    blocked_active = np.zeros((n_stick_y-1,n_stick_x-1))
    active_cell_mask = np.ones(mg.shape) # Can be used to mask regions of no-coverage
    # Determine active cells
    for i in np.arange(n_stick_x-1):
        for j in np.arange(n_stick_y-1):
            start_x=int(np.ceil(i*x_step))
            end_x=int(np.floor((i+1)*x_step))
            start_y=int(np.ceil(j*y_step))
            end_y=int(np.floor((j+1)*y_step))
            blocked_active[j,i] = np.any(active_area_full[start_y:end_y+1,start_x:end_x+1])
            active_cell_mask[start_y:end_y+1,start_x:end_x+1] = np.any(active_area_full[start_y:end_y+1,start_x:end_x+1])
    blocked_active = blocked_active.astype(bool)
    active_cell_mask = active_cell_mask.astype(bool)
    
    # Determine active sticks
    active_sticks = np.zeros((n_stick_y,n_stick_x)).astype(bool)
    for i in np.arange(n_stick_x):
        for j in np.arange(n_stick_y):
            if((i,j)==(0,0)): # Top left stick
                active_sticks[j,i] = blocked_active[j,i]
            elif ((i,j)==(n_stick_x-1,0)): # Top right stick
                active_sticks[j,i] = blocked_active[j,i-1] 
            elif ((i,j)==(0,n_stick_y-1)): # Bottom left
                active_sticks[j,i] = blocked_active[j-1,i]
            elif ((i,j)==(n_stick_x-1,n_stick_y-1)): # Bottom right
                active_sticks[j,i] = blocked_active[j-1,i-1]
            elif (j==0): # Top row
                active_sticks[j,i] = blocked_active[j,i] | blocked_active[j,i-1]
            elif (j==n_stick_y-1): # Bottom row
                active_sticks[j,i] = blocked_active[j-1,i] | blocked_active[j-1,i-1]
            elif (i==0): # left column
                active_sticks[j,i] = blocked_active[j-1,i] | blocked_active[j,i]    
            elif (i==n_stick_x-1): # right column
                active_sticks[j,i] = blocked_active[j-1,i-1] | blocked_active[j,i-1]   
            else: # centre cells
                active_sticks[j,i] = blocked_active[j-1,i-1] | blocked_active[j,i-1] | blocked_active[j,i] | blocked_active[j-1,i]
                
                
    if(init=='logit'):
    # Initiate in logit space            
        logit_low = sp.special.logit(low/10**6)
        logit_upp = sp.special.logit(upp/10**6)
        stick_heights = sp.special.expit((logit_upp - logit_low)*np.random.random((n_stick_y,n_stick_x)) + logit_low)*1000000 # Inititiate stick height array as random nums between 0 and 1000000mg/kg
    # Initiate in linear space
    elif (init=='linear'):
        print("Initiating grid in linear space...")
        stick_heights = np.random.random((n_stick_y,n_stick_x)) *1000000 # Inititiate stick height array as random nums between 0 and 1000000mg/kg
    
    inactive_cell_mask = np.invert(active_cell_mask) # Boolean array indicating which model grid cells are *not* covered 
     
    if(plot==True): # Visualises the inversion grid and covered region
        
        eg = z_to_grd(stick_heights)
        eg[np.where(coverage_mask)]='NaN'
        current_cmap = plt.cm.get_cmap()
        current_cmap.set_bad(color='grey')
        plt.imshow(eg,origin='lower')
        plt.colorbar()
        plt.ylabel("Vertical model grid cells")
        plt.xlabel("Horizontal model grid cells")
        plt.scatter(np.where(stick_heights)[1]*x_step,np.where(stick_heights)[0]*y_step,c='b',marker='x')  
        plt.scatter(np.where(active_sticks)[1]*x_step,np.where(active_sticks)[0]*y_step,c='r',marker='x') 
        plt.title("Active and inactive nodes (red/blue) of inversion grid with random values.")
    
    return((stick_heights,active_sticks))


(stix,active_stix) = initiate_inversion(40,34,plot=True,init='logit',upp=2000)
```

### Digression into expit, logit etc... 

We need to find a way to traverse the space between 0 and 1 efficiently. One option is to use the $expit$/$logit$ functions which map real values (from $-\infty$ to $\infty$) into the restricted domain 0 and 1 (and visa versa respectively). These $logit$ and $expit$ functions, and how they translate into mgkg are shown below. Note the clear sigmoid shape in linear space, but in logarithmic space, the relationship to concentration is linear. 

```python
logitx = np.linspace(-14,14,1000)
expit_mgkg = sp.special.expit(logitx)*1e6

plt.subplot(1,2,1)
plt.title('Concentration = 1e6 * expit(x)')
plt.plot(logitx,expit_mgkg)
plt.ylabel('Concentration mg/kg')
plt.xlabel('x')

plt.subplot(1,2,2)
plt.plot(logitx,expit_mgkg)
ax=plt.gca()
ax.set_yscale('log')
plt.ylabel('Concentration mg/kg')
plt.xlabel('x')

plt.show()
```

However, we need to be careful how we interpolate bilinearly between these values. Here we quickly visualise what the different interpolation methods look like so we can decide which of these methods should be preferred.

```python
n=20
low=1 # mg/kg
upp=999999 # mg/kg

np.random.seed(1)

random_seq = np.random.random(n)
raw_traverse = (upp-low) * random_seq + 1 # Random number between 1 and 999999 mg/kg
logit_traverse = (sp.special.logit(upp/1e6) - sp.special.logit(low/1e6))*random_seq + sp.special.logit(low/1e6)

stick_locs = np.linspace(1,n,n)
interp_locs = np.linspace(1,n,(n-1)*1000)

raw_interp = np.interp(interp_locs,stick_locs,raw_traverse)
expit_interp_nonlin = sp.special.expit(np.interp(interp_locs,stick_locs,logit_traverse))*1e6
expit_interp = np.interp(interp_locs,stick_locs,sp.special.expit(logit_traverse)*1e6)

plt.subplot(2,1,1)
plt.plot(interp_locs,raw_interp)
plt.scatter(stick_locs,raw_traverse)
plt.plot(interp_locs,expit_interp)
plt.plot(interp_locs,expit_interp_nonlin)
plt.scatter(stick_locs,sp.special.expit(logit_traverse)*1e6)
plt.legend(['Linear traverse and interp.','Logit traverse and linear interp.','Logit traverse and logit interp.'])
plt.ylabel('Concentration mg/kg')
plt.xlabel('Distance (between sticks)')
plt.title("Linear Scale")

plt.subplot(2,1,2)
plt.plot(interp_locs,raw_interp)
plt.scatter(stick_locs,raw_traverse)
plt.plot(interp_locs,expit_interp)
plt.plot(interp_locs,expit_interp_nonlin)
plt.scatter(stick_locs,sp.special.expit(logit_traverse)*1e6)
plt.legend(['Linear traverse and interp.','Logit traverse and linear interp.','Logit traverse and logit interp.'])
plt.ylabel('Concentration mg/kg')
plt.xlabel('Distance (between sticks)')
plt.yscale('log')
plt.title("Logarithmic Scale")

plt.tight_layout()
plt.show() 
# For each method have two side by side plots. 
# Each is 10 sticks chosen randomly from traversed interval and interpolated in the correct manner 
# Then visualise it both raw values (from 0 to 10000000mg/kg) and log values. 
```

The linear traverse and interpolation is the most intuitive, but a random sweep of this space does not cover low concentrations. In otherwords, it is not suited for log-normally distributed data. The $logit$ traverse *is* capable of identifying very low concentrations. This is good. The two interpolation methods are likely not too dissimilar. The only difference is that interpolation _after_ $expit$ is linear in concentration space, but interpolation _before_ is non-linear.  In addition, interpolation _after_ expit, may be more similar to the 'mixing' processes that occur in real life. i.e., a very low concentration region (10mg/kg) next to a high concentration region (100000mg/kg) would likely bleed into eachother via mixing, which is linear. Thus, we consider the second option ($logit$ traverse followed by $expit$ transform and then a linear interpolation) to be the most appropriate for our purpose. For parsimony we will also consider the simply linear traverse/interpolant for the time being too. 

It is worth noting here that using the $logit$ transform method removes the need for bounds on the inversion. The time cost of the $expit$ function is negligible. (200 $\mu$s for calculating expit of $10^5$ values)

We now define a function that, given certain parameters plots up a 2D view of the misfit surface, generated when 

```python
def visualise_misfit_space_2D(nx,ny,init_low=1,init_upp=999999,trav_low=1,trav_upp=999999,num_values=20,elem='Mg',stick1=(1,1),stick2=(2,2)):
    """ Visualises a 2D transect of the misfit space given appropriate parameters
    nx, ny = number of x and y sticks to initiate
    init_low/init_upp = lower/upper bounds for random values to initiate grid at. 
    trav_low/trav_upp = lower/upper bounds of parameter space to traverse across. 
    num_values = number of values between bounds to evaluate at (function is O(N^2)).
    elem = element to determine costfunction against 
    stick1/stick2 = node coordinates of sticks to vary height of. (NB: correct form if (ROW,COL))
    
    This automatically computes both the misfit space for a linear and logistic traverse."""
    
    # Initiative an inversion grid
    stix, active_stix = initiate_inversion(nx,ny,low=init_low,upp=init_upp,plot=True)
    
    # Select random nodes (currently at nodes 1 and 2)
    plt.scatter(x=[stick1[1]*x_step,stick2[1]*x_step],y=[stick1[0]*y_step,stick2[0]*y_step],s=250,facecolors='none', edgecolors='r')
    plt.text(x=stick1[1]*x_step,y=stick1[0]*y_step,s='Stick1')
    plt.text(x=stick2[1]*x_step,y=stick2[0]*y_step,s='Stick2')
    plt.title("Selected nodes for misfit exploration shown as circles")
    plt.show() 
    
    # Bounds for the space to traverse for inversion 
    low=trav_low
    upp=trav_upp

    # Transforming upper and lower bounds into logit space
    logit_low = sp.special.logit(low/10**6)
    logit_upp = sp.special.logit(upp/10**6)
    

    # Initiating grids and stick values
    raw_stick_values = np.linspace(low,upp,num_values)
    raw_misfit = np.zeros((num_values,num_values))
    raw_stix = np.copy(stix)

    logit_stick_values = np.linspace(logit_low,logit_upp,num_values)
    logit_misfit = np.zeros((num_values,num_values))
    logit_stix = np.copy(stix)

    # Looping through all possible stick values and evaluating the misfit
    # Each loop should cost ~ 50 ms 
    print("Predicted runtime = ",0.05*(num_values**2),"seconds")
    start = time.time()
    for i in np.arange(num_values):
        for j in np.arange(num_values):
            # Updating stick heights for raw traverse
            raw_stix[stick1] = raw_stick_values[i] # Stick 1
            raw_stix[stick2] = raw_stick_values[j] # Stick 2
            raw_interpd = z_to_grd(raw_stix) # Convert to continuous grid
            raw_misfit[j,i] = cost_homo(raw_interpd.reshape(flat_shape),elem) # Evaluate cost function

            # Updating stick heights for logit traverse
            logit_stix[stick1] = sp.special.expit(logit_stick_values[i])*1e6
            logit_stix[stick2] = sp.special.expit(logit_stick_values[j])*1e6
            logit_interpd = z_to_grd(logit_stix) # Convert to continuous grid
            logit_misfit[j,i] = cost_homo(logit_interpd.reshape(flat_shape),elem) # Evaluate cost function    
    end = time.time()
    print("Actual runtime=",end-start,"seconds")
    
    # Plot results for traverse
    plt.subplot(1, 2, 1)
    plt.imshow(raw_misfit,origin='lower',extent=[low,upp,low,upp])
    plt.colorbar()
    raw_min_loc = np.where(raw_misfit==np.amin(raw_misfit))
    raw_x_loc=raw_min_loc[1][0]
    raw_y_loc=raw_min_loc[0][0]
    plt.scatter(raw_stick_values[raw_x_loc],raw_stick_values[raw_y_loc],marker='+',color='r',s=200)
    plt.xlabel("Stick 1 value, mg/kg")
    plt.ylabel("Stick 2 value, mg/kg")
    plt.title("Data Misfit Map for Raw Traverse")
    
    plt.subplot(1, 2, 2)
    plt.plot(raw_stick_values,raw_misfit[raw_y_loc,:])
    plt.plot(raw_stick_values,raw_misfit[:,raw_x_loc])
    plt.legend(['Horizontal (Stick 1)','Vertical (Stick 2)'])
    plt.xlabel("Stick value mg/kg")
    plt.ylabel("Data misfit")
    plt.title("Transects along Raw Data Misfit through minimum")
    plt.tight_layout()
    plt.show() 

    # Plot results for logit traverse
    plt.subplot(1, 2, 1)
    plt.imshow(logit_misfit,origin='lower',extent=[logit_low,logit_upp,logit_low,logit_upp])
    plt.colorbar()
    logit_min_loc = np.where(logit_misfit==np.amin(logit_misfit))
    logit_x_loc=logit_min_loc[1][0]
    logit_y_loc=logit_min_loc[0][0]
    plt.scatter(logit_stick_values[logit_x_loc],logit_stick_values[logit_y_loc],marker='+',color='r',s=200)
    plt.xlabel("Stick 1 (logit) value")
    plt.ylabel("Stick 2 (logit) value")
    plt.title("Data Misfit Map for Logit Traverse")

    plt.subplot(1, 2, 2)
    plt.plot(logit_stick_values,logit_misfit[logit_y_loc,:])
    plt.plot(logit_stick_values,logit_misfit[:,logit_x_loc])
    plt.legend(['Horizontal (stick 1)','Vertical (stick 2)'])
    plt.xlabel("Stick value (logit)")
    plt.ylabel("Data misfit")
    plt.title("Transects along Logit Misfit through minimum")
    plt.tight_layout()
    plt.show() 
```

So far we have considered only the range 1 to 2000 mg/kg as this is the range of U in our study area according to the G-BASE survey (see above). We need to investigate what the misfit space looks like for this full range. We explore this using the same example above with a 10x10 grid.

We now initiate the grid across the full possible range of possible concentrations. Thus this examples represent the most generic case of the inversion. 

```python
visualise_misfit_space_2D(nx=10,ny=10,trav_low=1,trav_upp=999999,init_low=1,init_upp=999999,elem='U',stick1=(2,2),stick2=(5,4))
```

This most generic case, which due to random initiation is far from the minimum shows a clear issue: The misfit space is effectively bimodal. It has a high concentration end and a low concentration end. In this scenario, the cost function is minimised by turning 'off' the sticks. But, because of the limit between 0 and 1, there is no way to reduce the cost function further. Thus a flat area. This is very challenging for gradient descent. 

Note that the _exact_ same step shape is observed for a different element, e.g. Mg (see below). This is because the -5 to 5 logit range is the range at which the absolute (not logarithmic) range is most variable. Thus, we can see that _far_ from equilibrium, we simply have a bimodal 'steppy' misfit space, but closer to the minimum it can become more U-shaped. The problem with this is that before we reach a minimum, a gradient descent operator might get 'stuck' at any number of flat bits. Thus it would be desirable to add some convexity to this misfit space to prevent 'running off' the extremes. 

In addition, even close the minimum the misfit space is still not particularly well defined. As a result, a gradient descent operator could stop too early. This then needs to be conditioned.

### Discussion

Conclusions from the above experimentation: 
1. A $logit$ traverse produces much more reasonable looking misfit spaces than the linear traverse.
2. The misfit space has issues with convexity at extreme values and the bounds. 
3. The misfit space is dependent on the initialisation values. 
4. There do not appear to be local minima.
5. Far from a minimum, the inverse space is bimodal/stepped and simply turns nodes on or off.
6. Closer to the minimum, the inverse space is more U-shaped, but can still display flat spaces.

# Regularising/Damping the cost function

The above results indicate that the misfit space looks like it needs to be regularised somehow.  

Our first attempt involves penalising nodes for deviating too strongly from our prior estimate. This damps the model by discouraging values that deviate from the mean. We utilise the L2-norm for this purpose, meaning that we add to our cost function $\mu || C - C_{prior} ||$, where $\mu$ is a damping factor, the value of which will be determined empirically by visual inspection. 

From now on we reject the linear traverse and only focus on $logit$ stick values. 

```python
def l2_cost(active_stick_arr,elem):
    """Returns the l2 norm of the model relative to the prior. input in logit space"""
    l2norm = np.linalg.norm(active_stick_arr - prior_wtd_avg_logit[elem])
    return(l2norm)

stix,active_stix = initiate_inversion(5,5,low=prior_wtd_avg['Mg'],upp=prior_wtd_avg['Mg']*2,plot=True)
l2_cost(sp.special.logit(stix[active_stix]/1e6),'Mg')
```

## Misfit space of regularised misfit solution

Now we explore the misfit space of this damped inversion. First we define a new function which plots up the misfit space. 


```python
def visualise_damped_misfit_space_2D(nx,ny,mu_=0.0001,init_low=1,init_upp=999999,trav_low=1,trav_upp=999999,num_values=20,elem='Mg',stick1=(1,1),stick2=(2,2)):
    """ Visualises a 2D transect of the misfit space given appropriate parameters
    nx, ny = number of x and y sticks to initiate
    init_low/init_upp = lower/upper bounds for random values to initiate grid at. 
    trav_low/trav_upp = lower/upper bounds of parameter space to traverse across. 
    num_values = number of values between bounds to evaluate at (function is O(N^2)).
    elem = element to determine costfunction against 
    stick1/stick2 = node coordinates of sticks to vary height of. (NB: correct form if (ROW,COL))
    
    This automatically computes both the misfit space for a linear and logistic traverse."""
    
    # Initiative an inversion grid
    plt.rcParams["figure.figsize"] = (6,5)    
    stix, active_stix = initiate_inversion(nx,ny,low=init_low,upp=init_upp,plot=True)
    

    # Select random nodes (currently at nodes 1 and 2)
    plt.scatter(x=[stick1[1]*x_step,stick2[1]*x_step],y=[stick1[0]*y_step,stick2[0]*y_step],s=250,facecolors='none', edgecolors='r')
    plt.text(x=stick1[1]*x_step,y=stick1[0]*y_step,s='Stick1')
    plt.text(x=stick2[1]*x_step,y=stick2[0]*y_step,s='Stick2')
    plt.title("Selected nodes shown as circles")
    plt.show() 
    
    # Bounds for the space to traverse for inversion 
    low=trav_low
    upp=trav_upp

    # Transforming upper and lower bounds into logit space
    logit_low = sp.special.logit(low/10**6)
    logit_upp = sp.special.logit(upp/10**6)
    
    logit_stick_values = np.linspace(logit_low,logit_upp,num_values)
    logit_rms_misfit = np.zeros((num_values,num_values))
    logit_l2_norm = np.zeros((num_values,num_values))
    logit_stix = np.copy(stix)

    logit_prior = prior_wtd_avg_logit[elem]
    # Looping through all possible stick values and evaluating the misfit
    # Each loop should cost ~ 50 ms 
    print("Predicted runtime = ",0.026*(num_values**2),"seconds")
    start = time.time()
    for i in np.arange(num_values):
        for j in np.arange(num_values):
            
            # Updating stick heights for logit traverse
            logit_stix[stick1] = sp.special.expit(logit_stick_values[i])*1e6
            logit_stix[stick2] = sp.special.expit(logit_stick_values[j])*1e6
            
            logit_interpd = z_to_grd(logit_stix) # Convert to continuous grid
            
            logit_rms_misfit[j,i] = cost_homo(logit_interpd.reshape(flat_shape),elem) # Evaluate cost function 
            logit_l2_norm[j,i] = l2_cost(sp.special.logit(logit_stix[active_stix]/1e6),elem)*mu_ # Evaluate l2norm
            
    end = time.time()
    print("Actual runtime=",end-start,"seconds")
    
    total_cost_function = logit_l2_norm + logit_rms_misfit
    plt.rcParams["figure.figsize"] = (10,12)
    
    # Plot results for logit traverse
    plt.subplot(3, 2, 1)
    plt.imshow(logit_rms_misfit,origin='lower',extent=[logit_low,logit_upp,logit_low,logit_upp])
    plt.colorbar() 
    plt.vlines(x=logit_prior,ymin=logit_low,ymax=logit_upp,colors='gray',linestyle='dashed') # Add the prior as lines
    plt.hlines(y=logit_prior,xmin=logit_low,xmax=logit_upp,colors='gray',linestyle='dashed')
    raw_logit_min_loc = np.where(logit_rms_misfit==np.amin(logit_rms_misfit))
    raw_logit_x_loc=raw_logit_min_loc[1][0]
    raw_logit_y_loc=raw_logit_min_loc[0][0]
    plt.scatter(logit_stick_values[raw_logit_x_loc],logit_stick_values[raw_logit_y_loc],marker='x',color='m',s=200)
    plt.xlabel("Stick 1 (logit) value")
    plt.ylabel("Stick 2 (logit) value")
    plt.title("Data Misfit Map for Logit Traverse")

    plt.subplot(3, 2, 2)
    plt.plot(logit_stick_values,logit_rms_misfit[raw_logit_y_loc,:])
    plt.plot(logit_stick_values,logit_rms_misfit[:,raw_logit_x_loc])
    plt.legend(['Horizontal (stick 1)','Vertical (stick 2)'])
    plt.xlabel("Stick value (logit)")
    plt.ylabel("RMS misfit")
    plt.title("Transects along Data Misfit through minimum")
    plt.tight_layout()
    plt.show() 
    
    plt.subplot(3, 2, 3)
    plt.imshow(logit_l2_norm,origin='lower',extent=[logit_low,logit_upp,logit_low,logit_upp])
    plt.colorbar()
    plt.vlines(x=logit_prior,ymin=logit_low,ymax=logit_upp,colors='gray',linestyle='dashed') # Add the prior as lines
    plt.hlines(y=logit_prior,xmin=logit_low,xmax=logit_upp,colors='gray',linestyle='dashed')
    prior_logit_min_loc = np.where(logit_l2_norm==np.amin(logit_l2_norm))
    prior_logit_x_loc=prior_logit_min_loc[1][0]
    prior_logit_y_loc=prior_logit_min_loc[0][0]
    plt.xlabel("Stick 1 (logit) value")
    plt.ylabel("Stick 2 (logit) value")
    plt.title("Damping factor (model misfit) map for Logit Traverse")

    plt.subplot(3, 2, 4)
    plt.plot(logit_stick_values,logit_l2_norm[prior_logit_y_loc,:])
    plt.plot(logit_stick_values,logit_l2_norm[:,prior_logit_x_loc])
    plt.legend(['Horizontal (stick 1)','Vertical (stick 2)'])
    plt.xlabel("Stick value (logit)")
    plt.ylabel("RMS misfit")
    plt.title("Transects along damping factor (model misfit) through prior")
    plt.tight_layout()
    plt.show() 
    
    plt.subplot(3, 2, 5)
    plt.imshow(total_cost_function,origin='lower',extent=[logit_low,logit_upp,logit_low,logit_upp])
    plt.colorbar()
    plt.vlines(x=logit_prior,ymin=logit_low,ymax=logit_upp,colors='gray',linestyle='dashed') # Add the prior as lines
    plt.hlines(y=logit_prior,xmin=logit_low,xmax=logit_upp,colors='gray',linestyle='dashed')
    total_logit_min_loc = np.where(total_cost_function==np.amin(total_cost_function))
    total_logit_x_loc=total_logit_min_loc[1][0]
    total_logit_y_loc=total_logit_min_loc[0][0]
    plt.scatter(logit_stick_values[total_logit_x_loc],logit_stick_values[total_logit_y_loc],marker='+',color='r',s=200)
    plt.scatter(logit_stick_values[raw_logit_x_loc],logit_stick_values[raw_logit_y_loc],marker='x',color='m',s=200)
    plt.xlabel("Stick 1 (logit) value")
    plt.ylabel("Stick 2 (logit) value")
    plt.title("Total misfit for Logit Traverse")

    plt.subplot(3, 2, 6)
    plt.plot(logit_stick_values,total_cost_function[total_logit_y_loc,:])
    plt.plot(logit_stick_values,total_cost_function[:,total_logit_x_loc])
    plt.legend(['Horizontal (stick 1)','Vertical (stick 2)'])
    plt.xlabel("Stick value (logit)")
    plt.ylabel("RMS misfit")
    plt.title("Transects along total misfit function through minimum")
    plt.tight_layout()
    plt.show()
```

# Gradient descent optimisation

To perform an inversion we need to define a damped objective function to minimise. 
Our total damped objective function, for a suite of (logit) node heights, $\mathbf{x}$, and a damping factor $\mu$ is:

$F_{obj}[\mathbf{x}]= \overbrace{|| \log_{10}(\mathbf{C_{obs}}) - \log_{10}(\mathbf{C_{pred}}) ||_2^2}^{\mathrm{Data \ Misfit}} + \mu^2 \overbrace{|| \mathbf{x} - \mathbf{x_{prior}} ||_2^2}^{\mathrm{Model \ Misfit}} $


```python
def damped_objf(active_heights_logit,all_sticks,active_sticks,elem,mu_):
    """heights inputted logit space"""
    all_sticks[active_sticks] = active_heights_logit 
    interpolated = sp.special.expit(z_to_grd(all_sticks))*1e6 
    data_misfit = cost_homo(interpolated.reshape(flat_shape),elem)**2
    damping = (l2_cost(active_heights_logit,elem)*mu_)**2
    return(data_misfit+damping)
```

Now we can pass this objective function to different minimisers to find the optimum.

## Running an optimal inversion

Having convinced ourselves that this method is able to be optimised we turn our attention to ironing out the wrinkles. First we redefine a stream-lined optimal inversion initiation algorithm, and define a slightly faster cost function. These faster implementations of the cost function also use the log/exp vs logit/expit functions as they are significantly faster and for values <0.1 are effectively exactly the same.

```python
def initiate_inversion_slim(nx,ny,elem,plot=True):
    """ Initiates a model grid with 'nx' sticks along the x-axis and 
    'ny' sticks along y-axis. Defines global variables about the grid (e.g. spacing). 
    Automatically works out which of these nodes are sampled by our data-coverage.
    
    Initiates grid to have ___in log space___ all nodes to prior value
    """
    
    # Defines global variables based on grid resolution. Some of these are used by z_to_grd() and cost function
    global x_sticks, y_sticks,x_step,y_step,n_stick_x,n_stick_y,continuous_prior
    n_stick_x = nx
    n_stick_y = ny

    width = mg.shape[1] # number of x cells in topographic model grid
    height = mg.shape[0] # number of y cells in topographic model grid

    x_sticks = np.linspace(0,width,n_stick_x) # location in model x-dir of the sticks
    y_sticks = np.linspace(0,height,n_stick_y) # location in model y-dir of the sticks
    
    active_area_full= active_area.reshape(full_shape)
    x_step=width/(n_stick_x-1)
    y_step=height/(n_stick_y-1)

    blocked_active = np.zeros((n_stick_y-1,n_stick_x-1))
    active_cell_mask = np.ones(mg.shape) # Can be used to mask regions of no-coverage
    
    # Determine active cells
    for i in np.arange(n_stick_x-1):
        for j in np.arange(n_stick_y-1):
            start_x=int(np.ceil(i*x_step))
            end_x=int(np.floor((i+1)*x_step))
            start_y=int(np.ceil(j*y_step))
            end_y=int(np.floor((j+1)*y_step))
            blocked_active[j,i] = np.any(active_area_full[start_y:end_y+1,start_x:end_x+1])
            active_cell_mask[start_y:end_y+1,start_x:end_x+1] = np.any(active_area_full[start_y:end_y+1,start_x:end_x+1])
    blocked_active = blocked_active.astype(bool)
    active_cell_mask = active_cell_mask.astype(bool)
    
    # Determine active sticks
    active_sticks = np.zeros((n_stick_y,n_stick_x)).astype(bool)
    for i in np.arange(n_stick_x):
        for j in np.arange(n_stick_y):
            if((i,j)==(0,0)): # Top left stick
                active_sticks[j,i] = blocked_active[j,i]
            elif ((i,j)==(n_stick_x-1,0)): # Top right stick
                active_sticks[j,i] = blocked_active[j,i-1] 
            elif ((i,j)==(0,n_stick_y-1)): # Bottom left
                active_sticks[j,i] = blocked_active[j-1,i]
            elif ((i,j)==(n_stick_x-1,n_stick_y-1)): # Bottom right
                active_sticks[j,i] = blocked_active[j-1,i-1]
            elif (j==0): # Top row
                active_sticks[j,i] = blocked_active[j,i] | blocked_active[j,i-1]
            elif (j==n_stick_y-1): # Bottom row
                active_sticks[j,i] = blocked_active[j-1,i] | blocked_active[j-1,i-1]
            elif (i==0): # left column
                active_sticks[j,i] = blocked_active[j-1,i] | blocked_active[j,i]    
            elif (i==n_stick_x-1): # right column
                active_sticks[j,i] = blocked_active[j-1,i-1] | blocked_active[j,i-1]   
            else: # centre cells
                active_sticks[j,i] = blocked_active[j-1,i-1] | blocked_active[j,i-1] | blocked_active[j,i] | blocked_active[j-1,i]
                         
    stick_heights = np.zeros((n_stick_y,n_stick_x)) + prior_wtd_avg_log[elem] # initiate nodes to prior
    inactive_cell_mask = np.invert(active_cell_mask) # Boolean array indicating which model grid cells are *not* covered
    
    continuous_prior = np.zeros(mg.shape) + prior_wtd_avg[elem] # Define a continuous prior useful for speeding up inversion [in mg/kg!!!!]
     
    if(plot==True): # Visualises the inversion grid and covered region       
        eg =  z_to_grd(stick_heights)
        eg[np.where(coverage_mask)]='nan'
        current_cmap = plt.cm.get_cmap()
        current_cmap.set_bad(color='grey')
        plt.imshow(eg,origin='lower',vmin=prior_wtd_avg_log[elem]*0.95,vmax=prior_wtd_avg_log[elem]*1.05)
        cb = plt.colorbar()
        cb.set_label('Node value, log')
        plt.ylabel("Vertical model grid cells")
        plt.xlabel("Horizontal model grid cells")
        plt.scatter(np.where(stick_heights)[1]*x_step,np.where(stick_heights)[0]*y_step,c='b',marker='x')  
        plt.scatter(np.where(active_sticks)[1]*x_step,np.where(active_sticks)[0]*y_step,c='r',marker='x') 
        plt.title("Active and inactive nodes (red/blue) of inversion grid with random values (in log space)")   
    return((stick_heights,active_sticks))

def l2_cost_2(active_stick_arr,elem):
    """Returns the l2 norm of the model relative to the prior. input in log space"""
    l2norm = np.linalg.norm(active_stick_arr - prior_wtd_avg_log[elem])
    return(l2norm)

def damped_objf_faster(active_heights_log,all_sticks,active_sticks,elem,mu_):
    """a slightly faster version of the objective function.
    heights inputted in log space. 
    """
    all_sticks[active_sticks] = active_heights_log
    continuous_prior[mg_active_cell_bool] = np.exp(z_to_grd(all_sticks)[mg_active_cell_bool])*1e6    
    data_misfit = cost_homo(continuous_prior.reshape(flat_shape),elem)**2
    damping = (l2_cost_2(active_heights_log,elem)*mu_)**2
    return(data_misfit+damping) 
```

## Smoothing the solution 

For high resolution inversions, we may need to add smoothing to prevent 'spiky' over-fitting. To do this we penalise, initially, the first-derivative of the maps. This requires a further hyper-parameter, lamda which also must be calibrated. The function defined below calculates the magnitude of the gradient of of a given set of nodes. Note that this function assumes zero-gradient for any boundary nodes. 


$F_{obj}[\mathbf{x}]= \overbrace{|| \log_{10}(\mathbf{C_{obs}}) - \log_{10}(\mathbf{C_{pred}}) ||_2^2}^{\mathrm{Data \ Misfit}} + \mu^2 \overbrace{|| \mathbf{x} - \mathbf{x_{prior}} ||_2^2}^{\mathrm{Model \ Misfit}}  + \lamda^2 \overbrace{|| \nabla \mathbf{x}  ||_2^2}^{\mathrm{Roughness \ Misfit}} $


```python
def cost_roughness(stickz,active_stickz):
    """Returns l2 norm of roughness in both directions"""
    copy = np.copy(stickz)
    copy[np.invert(active_stickz)] = 'nan'
    padded = np.pad(copy,pad_width=1,mode='constant',constant_values='nan')
    grad=np.gradient(padded)
    rough_x = np.sqrt(np.nansum(grad[0]**2)) # Roughness in x direction (dC/dx)
    rough_y = np.sqrt(np.nansum(grad[1]**2)) # Roughness in y direction (dC/dy)
    return(rough_x,rough_y) # return tuple of roughness along both axes
```

Now we define an objective function that incorporates this damping term.

```python
def damped_smoothed_objf(active_heights_logit,all_sticks,active_sticks,elem,mu_,lamda_):
    """a slightly faster version of the objective function.
    heights inputted in logit space. 
    """
    all_sticks[active_sticks] = active_heights_logit
    continuous_prior[mg_active_cell_bool] = sp.special.expit(z_to_grd(all_sticks)[mg_active_cell_bool])*1e6    
    data_misfit = cost_homo(continuous_prior.reshape(flat_shape),elem)**2
    damping = (l2_cost(active_heights_logit,elem)*mu_)**2
    rough_x,rough_y = cost_roughness(all_sticks,active_sticks) 
    smoothing = (lamda_**2)*(rough_x**2 + rough_y**2)
    
    return(data_misfit+damping+smoothing)
```

Now we test this new damped objective function. 



```python
##### Set inversion parameters here ####

nx=10   # <<<<<<<<<<< Change values
ny=10  # <<<<<<<<<<< 
mu=0.1   # <<<<<<<<<<< 
lamda=0.1 # <<<<<<<<<<
elem='U'   # <<<<<<<<<<< 

### Initiating inversion ####

stix, active_stix = initiate_inversion_slim(nx=nx,ny=ny,elem=elem,plot=True) # Initiate inversion grid
plt.show()
active_array = np.copy(stix[active_stix]) # The array we will vary. 

#### Perform inversion ####

start = time.time()
print("Starting at: ", datetime.now())
sys.stdout = console_output # Switch to console output to print iteration updates there
res_nm = sp.optimize.minimize(fun=damped_smoothed_objf,args=(stix,active_stix,elem,mu,lamda),x0=active_array,method='Nelder-Mead',
                          options={'disp':True,'fatol':1e-4,'xatol':1e-4,'maxiter':2e7,'adaptive': True})
sys.stdout = nb_output # Switch to notebook output again
end = time.time()

#### Finish ####

print("############ results ############")
print("Finished at:", datetime.now())
print("Runtime = ",end-start,"s")
print(res_nm.success)
print(res_nm.status)
print(res_nm.message)
print(res_nm.nit)
smoothed_out=continuous_prior
```

## Evaluating inversion success

### Loading in independent G-BASE constraints 


Whilst the above inversion is, by definition, a good fit for the downstream data, we want to know whether it is a good fit for the 'target' G-BASE data. We thus turn our attention to the independent test which G-BASE offers.  

```python
def compare_to_gbase(inverse_results_grd,elem):
    """
    Compares the target inverse map to the independent G-BASE dataset. 
    Only needs to be provided with an element and a modelled grid and it 
    makes the comparison spatially, as a cross plot and shows misfit maps.
    """
    
    plt.rcParams["figure.figsize"] = (15,15)
    
    # Set upper and lower bounds
    invmin = np.nanmin(inverse_results_grd+coverage_nan_mask)
    invmax = np.nanmax(inverse_results_grd+coverage_nan_mask)
    
    # Plot the inverse results
    plt.subplot(3,2,1)
    plt.imshow(inverse_results_grd+coverage_nan_mask,origin='lower',norm=LogNorm(),vmin=invmin,vmax=invmax)
    cb = plt.colorbar()
    cb.set_label("Predicted concentration (mg/kg)")
    plt.xlabel("Horizontal Model Cells")
    plt.ylabel("Vertical Model Cells")
    plt.title("Inverse results")

    # Load in continuous G-BASE map
    tgt_comp_nc=netCDF4.Dataset('input_dir/input_chems/CG_log_'+elem+'.nc')
    tgt_comp_ma = tgt_comp_nc['z'][:,:]
    tgt_comp_bedrock = 10**tgt_comp_ma.data.astype(float).reshape(flat_shape)
    
    # Plot G-BASE map
    plt.subplot(3,2,2)
    plt.imshow(tgt_comp_bedrock.reshape(full_shape)+coverage_nan_mask,origin='lower',vmin=invmin,vmax=invmax,norm=LogNorm())
    cb = plt.colorbar()
    cb.set_label("G-BASE concentration (mg/kg)")
    plt.xlabel("Horizontal Model Cells")
    plt.ylabel("Vertical Model Cells")
    plt.title("G-BASE data")

    #### Extract G-BASE point observations and generate cross plot ####
    elem_rawdata =  np.loadtxt('input_dir/masked_'+elem+'.dat',dtype=str)
    good_records = np.invert(elem_rawdata[:,2] == 'NA') # Locations where not NA
    gbase_locs = elem_rawdata[good_records,:2].astype(float) # x, y
    gbase_locs_inds = np.flip(np.around(gbase_locs/mg.dx).astype(int),axis=1).T # indices. y then x
    gbase_indxs = list(zip(gbase_locs_inds[0,:],gbase_locs_inds[1,:])) # list of tuples of (y,x) indices
    
    results_grid = inverse_results_grd.reshape(full_shape).copy()
    predicted = np.log10(results_grid[gbase_locs_inds[0],gbase_locs_inds[1]])
    observed =  elem_rawdata[good_records,2].astype(float)
    misfit = predicted - observed
    
    # Plot data as a cross-plot
    plt.subplot(3,2,3)
    plt.scatter(x=predicted,y=observed,marker='.',s=0.5,color='darkgrey')
    plt.plot(np.unique(predicted), np.poly1d(np.polyfit(predicted, observed, 1))(np.unique(predicted)),c='k')
    abline(1,0)
    plt.xlabel('Inversion Predictions log10(mg/kg)')
    plt.ylabel(elem+'observations log10(mg/kg)')
    ax=plt.gca()
    ax.set_aspect('equal')
    ax.set_xlim(np.log10([invmin,invmax]))
    ax.set_ylim(np.log10([invmin,invmax]))
    plt.grid(True)

    # Output R2
    correlation_matrix = np.corrcoef(predicted, observed)
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2
    print("R-squared value:", r_squared)

    
    #### Plotting spatial misfit ####
    plt.subplot(3,2,4)
    plt.scatter(x=gbase_locs[:,0],y=gbase_locs[:,1],c=misfit,cmap='seismic',s=4,facecolor='grey',norm=DivergingNorm(vcenter=0))
    plt.gca().set_aspect('equal')
    plt.gca().set_facecolor('grey')
    cb = plt.colorbar()
    cb.set_label("(Underprediction)                  Misfit                  (Overprediction) ")
    
    ### Plotting distribution of misfits ####
    plt.subplot(3,2,5)
    plt.hist(x=misfit,bins=100)
    plt.xlabel('(Underprediction)                  Misfit                  (Overprediction) ')
    plt.tight_layout()
    plt.show()

compare_to_gbase(not_smoothed_out,'U')    
plt.show()
compare_to_gbase(smoothed_out,'U')
plt.show()
```

Finally we generate long profiles of geochemical composition for each of the major rivers alongside the fitted compositions.

```python
def compare_long_profiles(results_grid,elem):
    """ Produce long profiles for each major catchment's geochemistry
     overlain with the observations"""
    
    plt.rcParams["figure.figsize"] = (15,5)
    
    comp_bedrock = results_grid.reshape(flat_shape)
    # Calculate sediment composition
    a, sed_comp = find_drainage_area_and_discharge(mg.at_node['flow__upstream_node_order'], mg.at_node['flow__receiver_node'],runoff = comp_bedrock)  # composition but homogeneous erosion
    with np.errstate(divide='ignore'):
        with np.errstate(invalid='ignore'):     
            sed_comp_norm = sed_comp /q_homo_incis
    sed_comp_norm[q_homo_incis==0] = comp_bedrock[q_homo_incis==0]
    preds = sed_comp_norm[loc_nodes] # Extract predictions
    obs = obs_data[elem] # Extract observations
    misfit = np.log10(preds)-np.log10(obs)   
    misfitmin,misfitmax = np.amin(misfit),np.amax(misfit)
    
    plt.figure() 
    ax1 = plt.subplot(1,2,1) # Long profile figure
    ax2 = plt.subplot(1,2,2) # Map 
    
    riverlist = ['Spey','Tay','Dee','Don','Deveron']
    for river in riverlist:
        if(river=='Spey'):
            outlet=spey_outlet
        if(river=='Tay'):
            outlet=tay_outlet
        if(river=='Dee'):
            outlet=dee_outlet
        if(river=='Don'):
            outlet=don_outlet
        if(river=='Deveron'):
            outlet=deveron_outlet

        profiler = ChannelProfiler(mg,main_channel_only=True,outlet_nodes=[outlet]) 
        profiler.run_one_step() # Extract longest profile in each channel
        
        # Extract node IDs and stream distances
        prof_id = list(profiler.data_structure[outlet].values())[0]["ids"] # NB only works if main channel only extracted
        prof_distances = list(profiler.data_structure[outlet].values())[0]["distances"] # NB only works if main channel only extracted
        prof_xy = np.unravel_index(prof_id, full_shape) # Convert stream nodes into xy coordinates
        
        # Extract downstream predictions along long profile
        prof_geochem = sed_comp_norm[prof_id]
        # Extract observations along profile
        obs_prof_val = obs[np.isin(loc_nodes,prof_id)]
        obs_prof_misfit = misfit[np.isin(loc_nodes,prof_id)] # Calculate misfit 
        prof_loc_nodes = loc_nodes[np.isin(loc_nodes,prof_id)] # sample locs on profile
        obs_prof_x = np.zeros(obs_prof_val.size) # Extract stream distance for each locality 
        for i in np.arange(obs_prof_x.size): # loop is used to deal with duplicates
            obs_prof_x[i]= prof_distances[prof_id==prof_loc_nodes[i]]
            
        # Plot geochemical profile    
        ax1.plot(prof_distances/1e3,prof_geochem)
        ax1.scatter(obs_prof_x/1e3,obs_prof_val)#,c=obs_prof_misfit,s=60,cmap='seismic',vmin = misfitmin,vmax=misfitmax, norm=DivergingNorm(vcenter=0))   
        
        # Plot map of long profiles
        ax2.imshow(zr.reshape(full_shape)+landmask,cmap='terrain',origin='lower',norm=DivergingNorm(vcenter=0))
        ax2.scatter(prof_xy[1],prof_xy[0],s=2)
        ax2.scatter(x=fitted_locs[np.isin(loc_nodes,prof_id),0]/mg.dx, 
                y=fitted_locs[np.isin(loc_nodes,prof_id),1]/mg.dx,s=30,marker='x',c='black')
           

    # Tidy up figures
    ax1.set_xlabel('Distance to mouth / km')
    ax1.set_yscale('log')
    ax1.set_ylabel('Predicted downstream'+elem+', mg/kg')
    ax1.legend(labels=riverlist)
    ax2.set_xlabel('Horizontal Model Cells')
    ax2.set_ylabel('Vertical Model Cells')
    
    plt.tight_layout()
    plt.show()
```
